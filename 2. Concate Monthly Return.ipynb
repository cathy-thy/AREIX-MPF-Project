```python
import pandas as pd
import os 
from os import listdir
from os.path import isfile, join
```

```python
company ='Fidelity'
folderpath = "C:/Users/thy10/Desktop/AREIX/MPF Data/" + company+" csv/"
filelist = [f for f in listdir(folderpath) if isfile(join(folderpath, f))]
funds_monthly_list=[]#for all funds

for file in filelist:
    monthly_return =0
    filepath = folderpath + file
    readfile = pd.read_csv(filepath, index_col='Valuation Date', parse_dates=True)
    df1=pd.DataFrame(readfile, columns=["Unit NAV"])
    df1.sort_values(by=['Valuation Date'], inplace=True)
    
    #get the name of the file
    base = os.path.basename(filepath)
    labelname=os.path.splitext(base)[0] 
    
    if(company=='Allianz'):
        df1.index=pd.to_datetime(df1.index, format="%y%m%d").to_period('M') #for allianz
    else:
        df1.index=pd.to_datetime(df1.index, format="%m%d%y").to_period('M') #for fidelity and Investco
    
    #to get the first and last
    df_start=df1.groupby(df1.index).head(1)
    df_end=df1.groupby(df1.index).tail(1)
    
    df_fund_monthly = df_start
    
    for i in range(len(df_start)):
        if(df_end.iloc[i][0]==df_start.iloc[i][0]and i!=0):
            monthly_return = (df_end.iloc[i][0]-df_end.iloc[i-1][0])/df_end.iloc[i-1][0]
        else:
            monthly_return = (df_end.iloc[i][0]-df_start.iloc[i][0])/df_start.iloc[i][0]    

        df_fund_monthly.iloc[i]['Unit NAV'] = monthly_return
    
    df_fund_monthly.rename({"Unit NAV": labelname},axis = "columns", inplace=True) #rename column
    
    #add to the list
    funds_monthly_list.append(df_fund_monthly)

df_monthly = pd.concat(funds_monthly_list,axis=1)  
```

## output as csv

```python
df_monthly.to_csv(company+'_df_monthly.csv')
print(df_monthly)
```
